```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the JSON file
file_path = 'data/EI-result.json'
with open(file_path, 'r') as file:
    json_data = [json.loads(line) for line in file]

# Create a DataFrame
df_EI = pd.DataFrame(json_data)
```

```python
file_path = 'data/PI-result.json'
with open(file_path, 'r') as file:
    json_data = [json.loads(line) for line in file]

# Create a DataFrame
df_PI = pd.DataFrame(json_data)
```

```python
file_path = 'data/UCB-result.json'
with open(file_path, 'r') as file:
    json_data = [json.loads(line) for line in file]

# Create a DataFrame
df_UCB = pd.DataFrame(json_data)
```

```python
n = len(df_EI["top"])
iterations = list(range(1, n + 1))
```

```python
# Line plot
plt.plot(iterations, df_EI["top"], label= "Expected Improvement")
plt.plot(iterations, df_PI["top"], label= "Probability Improvement")
plt.plot(iterations, df_UCB["top"], label= "UCB Aquisition")


# Set labels and title
plt.xlabel('Itrations')
plt.ylabel('Top 1%')
plt.title("Performance for different aquisition functions with Bayesian Active Learning")
plt.legend()

# Show the plot
plt.show()
```
